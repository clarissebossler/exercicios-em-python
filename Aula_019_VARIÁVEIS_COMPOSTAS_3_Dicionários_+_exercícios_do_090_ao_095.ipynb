#DICIONÁRIOS - (aula 19)

Relembrando: Tuplas usam parenteses, Listas usam colchetes, Dicionários usam chaves:

*   Tuplas( )
*   Listas[ ]
*   Dicionários{ }   


Principal diferença entre listas e Dicionários:

*   Nas listas os índices são numéricos
*   Nos dicionários os índices são literais (são chamados de chaves, keys)

Partes de um dicionário:


*   valor (.values()): ---star wars---
*   chave (.keys()): ---filme---
*   item (.items()): ---filme: star wars---  

Dicionários podem estar dentro de listas, como na figura 4.

DICIONÁRIOS.jpg

```
#Declarando e vizualizando dicionários:
pessoas = {'nome': 'Gustavo', 'sexo': 'M', 'idade': 22}
print('A)', pessoas)
print('B)', pessoas['nome'])                     #usar aspas
print(f'C) O {pessoas["nome"]} tem {pessoas["idade"]} anos.')  #se ja tiver usado aspas, utilizar aspas duplas
print('D)', pessoas.keys())
print('E)', pessoas.values())
print('F)', pessoas.items(), '\nVerifique atravez dos símbolos (, [ e { do item F), que o dicionário é composto de uma lista com 3 tuplas.\n')

#Acessando os valores dos dicionarios atraves do for:
print('G)', end = '')
for k in pessoas.keys():
  print(k)

print('\nH)', end = '')
for k in pessoas.values():
  print(k)

print('\nI)', end = '')
for k in pessoas.items():    
  print(k)

#como o item tem 2 coisas (keys e values), pode-se usar 2 variáveis para utilizá-las separadamente (k, v) 
print()          #Não se utiliza o enumerate, e sim .items()
for k, v in pessoas.items():      #melhorando  a formatação:
  print(f'{k} = {v}')

#Substituindo, adicionando e apagando dados de um dicionário:
pessoas['nome'] = 'Leandro'    #substituindo
print('\nJ)', end = '')        
for k, v in pessoas.items():      
  print(f'{k} = {v}')


#del pessoas['sexo']
```

```
#Alimentando lista com dicionarios com o for:
#o dicionário estado armazena dados temposrariamete, em cada laço, que serão copiados no final do laco para a lista brasil.
estado = dict()
brasil = list()
for c in range(0,3):
  estado['uf'] = str(input('Unidade Federativa (nome do Estado): '))   #estado na posição 'uf' recebe.
  estado['sigla'] = str(input('Sigla do Estado: '))
  brasil.append(estado.copy())   #em vez de [:], os dicionários possuem um método chamado copy()

print('\nA)', brasil)

#melhorando a formatação do print:
print('\nB) Melhorando a formatação:')
for e in brasil:    #este for é para a lista
  for k, v in e.items():      #este for é para o dicionario
    print(f'O campo -{k}- tem valor -{v}-')

#melhorando a formatação do print:
print('\nC) Melhorando a formatação:')
for e in brasil:    #este for é para a lista
  for k, v in e.items():      #este for é para o dicionario
    print(f'{v}', end = ' G')
  print()
```

# EXERCÍCIO 090 - Dicionário em Python:

Desenvolva um programa que leia nome e média de um aluno, e que guarde também a situação em um dicionário (aprovado, recuperação, reprovado). No final, mostre o conteúdo da estrutura na tela.

https://www.youtube.com/watch?v=HipQYUk4koA&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=23

```
aluno = dict()
aluno['Nome'] = str(input('Nome do aluno: ')).strip().upper()
aluno['Média'] = float(input(f'Média de {aluno["Nome"]}: '))
if aluno['Média'] >= 7:
  aluno['Situação'] = 'Aprovado(a).'
elif 5 <= aluno['Média'] < 7:
  aluno['Situação'] = 'Em Recuperação.'
else:
  aluno['Situação'] = 'Reprovado(a).'
for k, v in aluno.items():
  print(f' - {k} é igual a {v}')
```

# EXERCÍCIO 091 - Jogo de dado utilizando dicionários:

Desenvolva um programa em que 4 jogadores joguem um dado e tenham resultados aleatórios. Guarde estes resultados em um dicionário. No final, coloque este dicionário em ordem, sabendo que o vencedor tirou o maior número no dado:

https://www.youtube.com/watch?v=cwrqIztaAwk&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=24

```
resultado = dict()
ranking = list()
from random import randint
from time import sleep
from operator import itemgetter    #"pega" um dado dentro de um item do dicionário (lembrando que item é key + value.)
for c in range(1, 5):
  resultado[f'jogador {c}'] = randint(1, 6)
for k, v in resultado.items():    #este FOR é para DICIONÁRIOS   (k, v, de key e value)
  print(f'O {k} tirou {v} no dado.')
  sleep(1)

print('{:=^53}'.format('RANKING'))
#repare que, ao usar o sorted, ele organiza em uma lista com tuplas. Logo, tratar ranking como lista.
ranking = sorted(resultado.items(), key=itemgetter(1), reverse=True)  #itemgetter: (0) para key, (1) para value
for i, v in enumerate(ranking):   #este FOR é para LISTAS   (i, de índice)
  sleep(1)
  print(f'O {i + 1}º colocado é o {v[0]}, com resultado {v[1]} no dado.')

print()
print(resultado)
print(ranking)
```

# EXERCÍCIO 092 - Dicionário que cresce ou diminui de acordo com o tipo de dado informado. (Cadastro de trabalhador)

Desenvolva um programa que leia nome, ano de nascimento e carteira de identidade e cadastre-os (com idade) em um dicionário. Se por acaso a CPTS for diferente de zero, o dicionário receberá também o ano de contratação e o salário. Calcule e acrescente, além da idade, com quantos anos a pessoa vai se aposentar:

https://www.youtube.com/watch?v=Vsqemzdrj78&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=25

```
from datetime import datetime
anoatual = datetime.now().year
trabalhador = dict()
trabalhador['Nome'] = str(input('Digite o nome do trabalhador: ')).strip().upper()
trabalhador['Idade'] = (anoatual - int(input('Digite o ano de nascimento: ')))
trabalhador['CTPS'] = int(input('Carteira de trabalho nº (0 para não tem): '))
if trabalhador['CTPS'] != 0:
  trabalhador['Ano de Contratação'] = int(input('Digite o ano de contratação: '))
  trabalhador['Salário'] = float(input('Salário: R$'))
  trabalhador['Aposentadoria'] = 35 - (anoatual - trabalhador['Ano de Contratação']) + trabalhador['Idade']
print('=-' * 20)
for k, v in trabalhador.items():
  print(f' - {k} tem o valor {v}.')
```

# EXERCÍCIO 093 - Cadastro de jogador de futebol:

Desenvolva um programa que gerencie o aproveitamento de um jogador de futebol. O programa vai ler o nome do jogador e quantas partidas ele jogou. Depois, vai ler a quatidade de gols feitos em cada partida. no final, tudo isso será guardado em um dicionário, incluindo o total de gols feitos: (apresentar os dados nos 3 métodos sugeridos pelo professor).

https://www.youtube.com/watch?v=5yKiud-YNaE&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=26

```
atleta = dict()
gols = list()  
tgols = partidas = 0
atleta['Nome'] = str(input('Nome do Jogador: ')).strip().upper()
partidas = int(input(f'Quantas partidas {atleta["Nome"]} jogou? '))
for p in range (1, partidas + 1):
  gols.append(int(input(f'Quantos gols {atleta["Nome"]} marcou na {p}ª partida? ')))
  tgols += gols[-1]      #Outra maneira de calcular o total de gols: método sum()  -ver linha 11.
atleta['Gols'] = gols
atleta['Total'] = tgols
#outra maneira de calcular o total de gols: atleta['Total'] = sum(gols)   -solução do professor.
print('=-' * 25)
print(atleta)

print('=-' * 25)
for k, v in atleta.items():
  print(f'O campo {k} tem valor {v}.')

print('=-' * 25)
print(f'O jogador {atleta["Nome"]} jogou {len(atleta["Gols"])} partidas.')
for p in range (1, partidas + 1): #ou: for i, v in enumerate(atleta['Gols']):
  print(f'  => Na partida {p}, fez {gols[p - 1]} gol(s).')
print(f'Foi um total de {tgols} gol(s).')
```

# EXERCÍCIO 094 - Unindo dicionários e listas:

Desenvolva um programa que leia nome, sexo e idade de várias pessoas, guardando os dados de cada pessoa em um dicionário e todos os dicionários em uma lista. No final, mostre: A) Quantas pessoas cadastradas. B) A média de idade. C) Uma lista com mulheres. D) Uma lista com idade acima da média.

https://www.youtube.com/watch?v=5yKiud-YNaE&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=26

```
pessoa = dict()
grupo = list()
somaidades = média = 0
while True:
  pessoa.clear()
  pessoa['nome'] = str(input('Nome: ')).strip().upper()
  while True:
    pessoa['sexo'] = str(input('Sexo [M/F]: ')).strip().upper()[0]
    if pessoa['sexo'] in 'MF':
      break
    print('Erro! Informe M ou F')
  pessoa['idade'] = int(input('Idade: '))
  somaidades += pessoa['idade']
  grupo.append(pessoa.copy())
  while True:
    cont = str(input('Deseja continuar? [S/N]')).strip().upper()[0]
    if cont in 'SN':
      break
    print('Erro! Informe S ou N')
  if cont == 'N':
    break

print('=-' * 30)
print(grupo)
print('=-' * 30)

#A)
print(f'A) Foram cadastradas {len(grupo)} pessoas.')

#B)
média = somaidades / len(grupo)
print(f'B) A média de idade do grupo é {média} anos.')

#C)
print(f'C) As pessoas do sexo feminino são:', end = ' ')
for p in grupo:
  if p['sexo'] == 'F':
    print(p['nome'], end = ' ')

#D)
print(f'\nD) As pessoas com idade acima da média são:')
for i, v in enumerate(grupo):      #nao precisaria usar o enumerate, pode ser como no item anterior (linha 36).
  if grupo[i]['idade'] > média:
    print(f'{grupo[i]["nome"]}, com {grupo[i]["idade"]} anos.')
print('\n<<ENCERRADO>>')
```

# EXERCÍCIO 095 - Aprimorando os dicionários:

Aprimore o exercício 093, para que ele funcione com vários jogadores, incluindo um sistema de vizualização de detalhes do aproveitamento de cada jogador.

https://www.youtube.com/watch?v=mw1So0r317Y&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=28

```
time = list()  
atleta = dict()
gols = list()
cont = ' '

while True:
  atleta.clear()
  gols.clear()
  tgols = partidas = 0
  atleta['Nome'] = str(input('Nome do Jogador: ')).strip().upper()
  partidas = int(input(f'Quantas partidas {atleta["Nome"]} jogou? '))
  for p in range (1, partidas + 1):
    gols.append(int(input(f'Quantos gols {atleta["Nome"]} marcou na {p}ª partida? ')))
    tgols += gols[-1]      #Outra maneira de calcular o total de gols: método sum()  -ver linha 11.
  atleta['Gols'] = gols[:]
  atleta['Total'] = tgols
  time.append(atleta.copy())
  while True:
    cont = str(input('Deseja continuar? [S/N] ')).strip().upper()[0]
    if cont in 'SN':
      break
    print('Erro! Digite S ou N')
  if cont == 'N':
    break

print('=-' * 20)
print(time)
print('=-' * 20)
print(' cód nome            gols        total')
print('-' * 40)     

#na solução das proximas 2 linhas, não é possível alinhar todos os dados da tabela.
#for t in range(0, len(time)):      
#  print(f'{t + 1:>3} {time[t]["Nome"]:<12} {time[t]["Gols"]} {time[t]["Total"]}')

#alinhando todos os dados da tabela:                 
for k, v in enumerate(time):           # na lista time, k é o índice, que representa o que está dentro das chaves. v é o que está dentro das chaves.
  print(f'{k + 1:>4} ', end = '')      # printa o código.
  for d in v.values():                 # dentro de v (que no caso é o conteúdo das chaves), o método values() pega só os valores: CARLOS, [1, 3, 0], 4 (que serão o d)
    print(f'{str(d):<15}', end = '')   #printa o d, e por alguma razao interna, printa alinhado.
  print()
print('=-' * 23)

while True:
  busca = int(input('Ver dados de qual jogador? (999 para parar): '))
  if busca == 999:
    break
  if busca >= len(time)+1:
    print('Código de jogador inexistente')
  else:
    print(f'---LEVANTAMENTO DO JOGADOR {time[busca - 1]["Nome"]}')
    for i, v in enumerate(time[busca - 1]['Gols']):    #i é o rank, v é o valor dos parênteses
      print(f'     No jogo {i+1} fez {v} gols.')
    print('=-' * 23)
print('>>FINALIZADO<<')
```
