# INTERROMPENDO REPETIÇÕES WHILE (break, True)
# STOPPING REPETITIONS *WHILE*
*   O comando "break" interrompe um laço EM QUALQUER PONTO (considerar a posição da linha com o comando BREAK)
*   Bom para utilizar com "flag".

BREAK.jpg

```
#Looping infinito:
n = 0
s = 0
while True:                                #O default é True, até que algo seja definido como False.
  n = int(input('Digite um número '))
  s += 1
print('A soma vale {}'.format(s))          #Esta linha não é alcançada, pois o programa nunca sairá do while.
```

```
#Utilizando o break COM o flag:
n = s = 0
while True:                                
  n = int(input('Digite um número '))
  if n == 999:        #999 é a flag
    break             #Atenção para a posição do break dentro do while: neste caso, fica antes da soma.
  s += n              #ATENÇÃO += é diferente de =+
print('A soma vale {}'.format(s)) 
```

# EXERCÍCIO 066 - Inserir vários números e usar FLAG para encerrar - *Input several numbers and use FLAG to end:*
 
Desenvolva um programa que leia vários números inteiros pelo teclado. O programa encerra quando o usuário digita 999, que é a CONDIÇÃO DE PARADA (flag). No final, o programa mostra quantos números foram digitados e o resultado da soma entre eles (desconsiderando o flag):
 
*Develop a program that reads several integer by the keyboard. It ends when the user types the flag 999, (stop condition). Them the program displays how  many numbers were entered and the result of the sum of all of them:*
 
https://www.youtube.com/watch?v=d2ug6quC1bk&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=36

```
print('Digite números para apresentar a soma entre eles')
soma = número = contador = 0     #Se utilizar looping infinito, não há necessidade do "número = 0" nesta linha.
while número != 999:             #O professor utilizou looping infinito (while True:)
  número = int(input('Digite um número (999 para encerrar): '))
  if número == 999:
    break
  soma += número
  contador += 1
print(f'Total de números: {contador}. Soma total entre os números: {soma}')
```

# EXERCÍCIO 067 - Tabuada v3.0 - *Multiplication table:*
 
Desenvolva um programa que mostre a tabuada de vários números (um por vez) para cada valor digitado pelo usuário. O programa encerra quando nele é inserido um número negativo:
 
*Develop a program that displays the multiplication table of several numbers (one at a time) for each value  typed by the user. Program ends when a negative number is typed:*
 
https://www.youtube.com/watch?v=X0a5aZg93Uc&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=37

```
print('Gerador de Tabuada v3.0')
print('-' * 42)
while True:
  tabuada = int(input('Insira um número para ver sua tabuada: '))
  print('-' * 42)
  if tabuada < 0:
    break
  for cont in range (1, 11):
    print(f'{cont} x {tabuada} = {cont * tabuada}')
print('FIM')
```

# EXERCÍCIO 068 - Jogo do par ou ímpar - *Even or odd game*:
Desenvolva um programa que jogue "par ou ímpar" com o usuário. O jogo encerra quando o usuário perde. Então, exibe o total de vitórias consecutivas que o usuário conquistou:
 
*Develop a program that plays Even Or Odd Game with the user. The game ends when the user loses. It then displays the total consecutive wins that the user has won:*
 
https://www.youtube.com/watch?v=EIzgKCCDdc0&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=38

```
print('=-' * 24)
print('          JOGO DO PAR OU ÍMPAR')
print('=-' * 24)
cont = 0
piusuário = ' '
from random import randint
while True:
  nusuário = int(input('Digite um número: '))
  ncomputador = randint(0, 10)
  piusuário = str(input('Par ou Ímpar? [P/I]: ')).strip().upper()[0]
  jogo = nusuário + ncomputador
  if piusuário == 'P':               #PARA JOGADOR ESCOLHEU PAR:
    if jogo % 2 == 0:   
      print(f'Você jogou {nusuário}, o computador escolheu {ncomputador}. Total = {jogo}')
      print('Deu PAR! Você GANHOU.')
    else:
      print(f'Você jogou {nusuário}, o computador escolheu {ncomputador}. Total = {jogo}')
      print('Deu ÍMPAR, Você PERDEU')
      break
  elif piusuário == 'I':              #PARA JOGADOR ESCOLHEU ÍMPAR:
    if jogo % 2 != 0:            
      print(f'Você jogou {nusuário}, o computador escolheu {ncomputador}. Total = {jogo}')
      print('Deu ÍMPAR! Você GANHOU.')
    else:
      print(f'Você jogou {nusuário}, o computador escolheu {ncomputador}. Total = {jogo}')
      print('Deu PAR! Você PERDEU')
      break
  else:
    print('Jogada inválida.')
    break
  cont += 1
  print('Vamos jogar novamente...')
  print('=-' * 24)
print(f'GAME OVER! Você ganhou {cont} vez(es) consecutiva(s)!')
print('-=' * 24)
```

```
#Versão com considerações do professor (2 whiles):
print('=-' * 24)
print('           JOGO DO PAR OU ÍMPAR')
print('=-' * 24)
cont = 0
from random import randint
while True:
  nusuário = int(input('Digite um número: '))
  ncomputador = randint(0, 10)
  jogo = nusuário + ncomputador
  piusuário = ' '
  while piusuário not in 'PI':      #este while continua perguntando até o usuário responder corretamente P ou I. também reduz as linhas de comando com a frase "voce escolheu x, o computador escolheu y..."
    piusuário = str(input('Par ou Ímpar? [P/I]: ')).strip().upper()[0]
  print(f'Você jogou {nusuário}, o computador escolheu {ncomputador}. \nTotal = {jogo}', end = '  -->  ')
  print('!DEU PAR!' if jogo % 2 == 0 else '!DEU ÍMPAR!')
  if piusuário == 'P':               #PARA JOGADOR ESCOLHEU PAR:
    if jogo % 2 == 0:
      print('VOCÊ VENCEU!')
      cont += 1
    else:
      print('VOCÊ PERDEU')
      break
  else:                              #PARA JOGADOR ESCOLHEU ÍMPAR:
    if jogo % 2 != 0:
      print('VOCÊ VENCEU!')
      cont += 1
    else:
      print('VOCÊ PERDEU')
      break
  print('Vamos jogar novamente...')
  print('=-' * 24)
print(f'GAME OVER! Você ganhou {cont} vez(es) consecutiva(s)!')
print('-=' * 24)
```

# EXERCÍCIO 069 - Análise de dados de um grupo de pessoas - *Analysis of goup data:*
 
Desenvolva um programa que leia a idade e o sexo de várias pessoas. Para cada pessoa cadastrada, o programa deve perguntar ao usuário se ele deseja continuar ou não. Ao final desde registro, o programa exibe: A) Quantas pessoas tem mais de 18 anos. B) Quantas pessoas do sexo feminino foram cadastradas. C) Quantas pessoas do sexo masculino têm menos de 20 anos:
 
*Develop a program that reads the age and gender of several people. For each registered person, it musts ask if the user wants to continue or not. At the end of this registration, the program  displays: A) How many people are over 18. B) How many female people were registered. C) How many male people are under 20.
 
https://www.youtube.com/watch?v=4Ca6iRJo3M0&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=39

```
print('CADASTRO DO GRUPO DE VIAGEM')
c1 = c2 = c3 = 0
while True:
  idade = int(input('Digite a idade: '))
  if idade <= 18:
    c1 += 1
  sexo = ' '
  while sexo not in 'MF':
    sexo = str(input('Digite o sexo: [M/F] ')).strip().upper()[0]
  if sexo == 'F':
    c2 += 1
  elif sexo == 'M' and idade < 20:
    c3 += 1
  continuar = ' '
  while continuar not in 'SN':
    continuar = str(input('Deseja continuar? [S/N] ')).strip().upper()[0]
  if continuar is 'N':
    break
print('=-' * 26)
print(f'Foram cadastradas {c1} pessoas com menos de 18 anos')
print(f'{c2} pessoas são do sexo feminino. {c3} pessoas são do sexo masculino com menos de 20 anos de idade.')
```

# EXERCÍCIO 070 - Estatísticas em produtos - *Product Statistics:*
Desenvolva um programa que leia o nome e o preço de vários produtos. Para cada produto cadastrado, ele deve perguntar ao usuário se ele deseja continuar ou não. Ao final desse registro, o programa exibe: A) o valor total da compra. B) quantos produtos custam mais de R$1000. C) o nome do produto mais barato:
 
*Develop a program that reads the name and price of several products. For each registered product, it musts ask if the user wants to continue or not. At the end of this registration, the program  displays: A) total purchase amount. B) how many products cost more then R$ 1000. C) The name of the cheapest product:*
 
https://www.youtube.com/watch?v=hS8QdW-1HTo&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=40

```
print('-=' * 25)
print('{:^50}'.format('LOJA SUPER BARATÃO'))
print('=-' * 25)
total = menor = c1 = c2 = 0
barato = ' '
while True:
  produto = str(input('Nome do produto: ')).strip().upper()
  preço = float(input('Valor do produto: R$ '))
  total += preço
  c1 += 1
  if c1 == 1:
    menor = preço
  else:
    if menor > preço:
      barato = produto
  if preço >= 1000:
    c2 += 1
  continuar = ' '
  while continuar not in 'SN':
    continuar = str(input('Deseja continuar? [S/N] ')).strip().upper()[0]
  if continuar == 'N':
    break
print('-=' * 25)
print(f'Total gasto: R${total:0.2f}\nQtde de produtos que custam mais de R$1000: {c2}\nProduto mais barato: {barato}.')
```

# EXERCÍCIO 071 - Simulador de caixa eletrônico - *ATM simulator*:
Desenvolva um programa que simule o funicionamento de um caixa eletrônico com as funcionalidades: A) ler o valor a ser sacado. B) exibir quantas cédulas de cada valor serão entregues (considerar cédulas de 50, 20, 10 e 1):

*Develop a program that simulates a ATM with the features: A) to read the amount to be withdraw. B) displays how many banknotes (of each type) will be delivered (condiser 50, 20, 10 and 1 banknotes types):*
 
https://www.youtube.com/watch?v=_XGgwltYpYk&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=41

```
print('=-' * 16)
print('{:^32}'.format('SAQUE CAIXA ELETRÔNICO'))
print('=-'* 16)
c50 = c20 = c10 = c1 = 0
valor = int(input('Qual valor você deseja sacar? (cédulas disponíveis: 50, 20, 10 e 1) R$ '))
while valor >= 50:
  valor -= 50
  c50 += 1
while valor >= 20:
  valor -= 20
  c20 += 1
while valor >= 10:
  valor -= 10
  c10 += 1
while valor >= 1:
  valor -= 1
  c1 += 1
print(f'Você receberá {c50} notas de R$50, {c20} notas de R$20, {c10} notas de R$10 e {c1} notas de R$1.')
#O problema desta versão é que pode aparecer "0 notas de R$10", por exemplo. (verificar se é possivel eliminar mantendo a estruruta do codigo)
```

```
#Solução do professor:
print('=-' * 16)
print('{:^32}'.format('SAQUE CAIXA ELETRÔNICO'))
print('=-'* 16)
valor = int(input('Qual valor você deseja sacar? (cédulas disponíveis: 50, 20, 10 e 1) R$ '))
total = valor
céd = 50
totcéd = 0
while True:
  if total >= céd:
    total -= céd
    totcéd += 1
  else:
    if totcéd > 0:
      print(f'O total de cédulas de R$ {céd} é {totcéd}.')
    if céd == 50:
      céd = 20
    elif céd == 20:
      céd = 10
    elif céd == 10:
      céd = 1
    totcéd = 0
    if total == 0:
      break
```
