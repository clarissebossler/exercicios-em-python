#VARIÁVEIS COMPOSTAS 1: TUPLAS

*   Os 3 tipos de variáveis compostas são: Tuplas, Listas e Dicionários.
*   Tuplas utilizam ( ), Listas [  ], e Dicionários { }
*   Tupla é uma variável capaz de receber mais de um item. Cada elemento é identificado por um índice (que é um número que se inicia em 0)

*   Para acessar os itens dentro da variável: colchetes, dentro deles número do índice.

*   Len (de lenth, comprimento) -> mostra quantos itens há dentro de uma variável.

*   Ao ser utilizado com o for, o "range" é a variável composta, e o "c" é cada item de cada vez (um em cada laço)

*   AS TUPLAS SÃO IMUTÁVES. Uma vez definidas, elas não podem ter seus itens substituídos durante a execução do programa.






TUPLA.jpg

```
print('{:-^9}'.format ('A'))
lanche = ('Hamburguer', 'Suco', 'Pizza', 'Pudim')
print(lanche)
#também funcionam sem os parênteses (a partir de determinada versao do Python)
lanche = 'Hamburguer', 'Suco', 'Pizza', 'Pudim'
print(lanche)

print('{:-^9}'.format ('B'))
#para referenciar o índice, sempre colchetes:
print(lanche[0])   #primeiro item, começando com 0
print(lanche[3])   #terceito item, começando com 0
print(lanche[-1])  #números negativos: de trás para frente, começando com -1
print(lanche[1:3]) #fatiamento do 1 até o 3 (ignorando o último)
print(lanche[2:])  #fatiamento do 2 até o 
print(lanche[:2])  #fatiamento do início até o segundo item (ignorando o último)
print(lanche[-2:]) #fatiamento do -2 até o final (pudim)

print('{:-^9}'.format ('C'))
#com FOR:
#A) "for normal", utilizando o range:
for c in range (0, len(lanche)):
  print(f'Eu vou comer {lanche[c]}')
#B) for diretamente da variável composta:
print('\n')
for itenscomestíveis in lanche:
  print(f'Eu vou comer {itenscomestíveis}')
#C) indicando a posição:
print('\n')
for pos, itenscomestíveis in enumerate(lanche):
  print(f'Em {pos}º lugar, vou comer {itenscomestíveis}')
7
print('{:-^9}'.format ('D'))
#mostrar os itens em ordem alfabética com SORTED:
print(sorted(lanche))
```

```
#uso do "juntar" + em tupla (+). A ordem influencia.
a = (2, 100, 4)
b = (100, 8, 1, 2)
c = b + a
print (c)

#len: conta quantos itens há na tupla, iniciando no 1.
print(f'A tupla c tem {(len(c))} itens.')

#count: conta algum item específico:
print(f'O número cinco aparece {(c.count(5))} vezes na tupla c.')

#index: mostra em qual posição (índice) está algum ítem:
print('Em qual índice está o número 8?', end = ' ')
print(c.index(8))

print('Em qual índice está o número 100?', end = ' ')
print(c.index(100))    #se houver mais de um item igual, mostra apenas o índice (posição) da primeira ocorrencia.

print(f'e também no índice {(c.index(100, 1))}.')  #este 1 significa: ler a partir da posição 1 (isto se chama deslocamento).
```

# EXERCÍCIO 072 - Número por extenso:
 
Desenvolva um programa que tenha uma tupla totalmente preenchida com uma contagem por extenso, de zero até vinte. Seu programa deverá ler um número pelo teclado e mostrá-lo por extenso.
 
https://www.youtube.com/watch?v=ei2Kr3ccfO0&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=2

```
extenso = ('Zero', 'Um', 'Dois', 'Três', 'Quatro', 'Cinco', 'Seis', 
           'Sete', 'Oito', 'Nove', 'Dez', 'Onze', 'Doze', 'Treze', 
           'Catorze', 'Quinze', 'Dezesseis', 'Dezessete', 'Dezoito', 
           'Dezenove', 'Vinte')
while True:         #consegui fazer com a validação, porém nao sei se esses "2 while true " seguidos são a melhor solução (ou se é estranho)
  while True:
    request = int(input('Digite um número entre 0 e 20 para vê-lo por extenso: '))
    if 0 <= request <= 20:
      print(f'Você digitou: -{extenso[request]}-')
      break
    else:
      print('Tente novamente. ', end = '')
  continuar = ' '
  while continuar not in 'SN':
    continuar = str(input('Deseja continuar? [S/N] ')).strip().upper()[0]
  if continuar == 'N':
    break
print('FIM')
```

# EXERCÍCIO 073 - Tuplas com times de futebol:
 
Crie uma tupla preenchida com os 20 primeiros colocador na Tabela do Campeonato Brasileiro de Futebol, na ordem de colocação. Depois, mostre:
 
A) Os 5 primeiros. B) Os últimos 4 colocados. C) Times em ordem alfabéica. D) Em que posição está o Chapecoense.
 
https://www.youtube.com/watch?v=RexybLcGewA&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=3

```
times = ('Corinthians', 'Palmeiras', 'Santos', 'Grêmio', 
         'Cruzeiro', 'Flamengo', 'Vasco', 'Chapecoense', 
         'Atlético', 'Botafogo', 'Atlético-PR', 'Bahia',
         'São Paulo', 'Fluminense', 'Sport Recife',
         'EC Vitória', 'Coritiba', 'Avaí', 'Ponte Preta', 
         'Atlético-GO')
print('{:=^90}'.format('TABELA DO CAMPEONATO BRASILEIRO DE FUTEBOL'))
for t in times: 
  print(t)
print(f'\nOs 5 primeiros colocados são: {times[:5]}\n')
print(f'Os últimos 4 colocados são: {times[-4:]}\n')
print(f'Os times em ordem alfabética: {sorted(times)}\n')
print(f'O Chapecoense está na {times.index("Chapecoense")+1}ª posição.')
```

# EXERCÍCIO 074 - Maior e menor valores em Tuplas:
 
Desenvolva um programa que vai gerar cinco números aleatórios e colocá-los em uma tupla. Depois disso, mostre a listagem de números gerados e também indique o menor e o maior valor que estão na tupla:
 
https://www.youtube.com/watch?v=mlwt2CRQkTQ&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=4

```
from random import randint
sorteio = (randint(1, 10), randint(1, 10), randint(1, 10), randint(1,10), randint(1, 10))
print(f'Os valores sorteados foram: ', end = '')
for n in sorteio:
  print(f'{n}', end = ' ')
print(f'\nO maior valor sorteado foi {max(sorteio)}')
print(f'O menor valor sorteado foi {min(sorteio)}')
```

# EXERCÍCIO 075 - Análise de dados em uma Tupla:
 
Desenvolva um programa que leia quatro valores pelo teclado e armazene-os em uma tupla. Então, mostre: A) Quantas vezes apareceu o valor 9. B) Em qual posição apareceu pela primeira vez o valor 3. C) Quais foram os números pares:
 
https://www.youtube.com/watch?v=1u7oA8ckjAc&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=5

```
número = 0
número = (int(input('Digite um número: ')), int(input('Digite outro número: ')), int(input('Digite mais um número: ')), int(input('Digite o último número: ')))
print(f'Você digitou os valores {número}')
print(f'O número 9 apareceu {número.count(9)} vez(es)')
if 3 in número:                                              #este IF é mecessário pois, caso não haja 3 na tupla, dará erro.
  print(f'O número 3 apareceu pela primeira vez na posição {(número.index(3)+1)}')
print('Os números pares digitados foram: ', end = '')
for item in número:
  if item % 2 == 0:
    print(f'{item}', end = ' ')
```

# EXERCÍCIO 076 - Alinhamento em Tupla (Lista de preços):
 
Desenvolva um programa que tenha uma tupla única com nomes de produtos e seus respectivos preços em sequência. Mostrar uma listagem de preços com estes dados organizados de forma tabular.
 
https://www.youtube.com/watch?v=Qp2cXfCHk2I&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=6

```
print('-' * 30)
print('{:^30}'.format('LISTA DE MATERIAIS'))
(print('-' * 30))
materiais = ('Lápis', 1.75,
             'Borracha', 2,
             'Caderno', 15.90,
             'Estojo', 25,
             'Transferidor', 15.90,
             'Compasso', 9.99,
             'Mochila', 120.00,
             'Canetas', 22.30,
             'Livro', 34.90)
for pos in range(0, len(materiais)):     #utilizando esta variável (pos), podemos trabalhar nela 
  if pos % 2 == 0:                       #posições pares serão sempre o NOME dos itens
    print(f'{materiais[pos]:.<20}', end = '')
  else:                                  #else, ou seja, se pos for ímpar. Neste caso, posicao ímpar sempre será PREÇO.
    print(f' R$ {materiais[pos]:>6.2f}')
print('-' * 30)
```

#EXERCÍCIO 077 - Contando vogais em Tupla:
 
Desenvolva um programa que tenha uma tupla com várias palavras (não usar acentos). Mostrar cada palavra e quais são suas vogais:
 
https://www.youtube.com/watch?v=8BgSqrOpKvU&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=7

```
palavras = ('aprender', 'programar', 'linguagem', 'python',
            'curso', 'gratis', 'estudar', 'praticar', 'trabalhar',
            'mercado', 'programador', 'futuro')
for item in palavras:
  print(f'\nNa palavra {item.upper()} temos as vogais', end = ' ')
  for letra in item:
    if letra.lower() in 'aeiou':
      print(letra.upper(), end = ' ')
```
