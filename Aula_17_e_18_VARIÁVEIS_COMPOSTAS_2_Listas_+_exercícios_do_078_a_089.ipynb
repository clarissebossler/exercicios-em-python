#LISTAS - (aula 17)

*   Relembrando: Tuplas usam parenteses, Listas usam colchetes, Dicionários usam chavez:

      Tuplas( )

      Listas[ ]

      Dicionários{ }

*   A Principal diferença entre tuplas e listas, é que as listas aceitam modificações e adições em seus itens (tanto no final, como entre os itens)e também apagar elementos, durante a execução do programa.

*   É possível criar listas atrévez do comando range



```
#Declaração de listas:
preços = []    #ou: notas = list()  Uma lista pode ser ou estar vazia de itens.
print(preços)
preços = [22, 44, 66, 88]    #listas são colocadas ENTRE COLCHETES
print('A)', preços)
print('B)', preços[2])    #mostra apenas o item indicado na posição (no caso, posição 0)

#Manipulação de listas:
preços[1] = 88     #significa: posição 2 recebe 88 (vai substituir o item preexitstente (44) por 99.
print('C)', preços)
preços.append(99)  #no final da lista, adiciona mais um valor (no caso, 99)
print('D)', preços)
preços.insert(1, 33)   #insere novos valores no meio da lista. (significa: na posição 1, inserir o valor 33)
print('E)', preços)
preços.sort()      #modifica a posição dos itens dentro da lista (organiza em ordem alfabética para letras, crescente para números)
print('F)', preços)
preços.sort(reverse=True)   #idem ao anterior, mas organiza em ordem descrescente.
print('G)', preços)
print(f'Esta lista tem {len(preços)} elementos')

#Apagando itens de uma lista:
preços.pop()      #elimina o ultimo item da lista
print('H)', preços)
preços.pop(4)     #elimina o ítem da posição 4 (e atualiza o índice dos itens, de acordo com as novas posições)
print('I)', preços)
preços.remove(88)    #elimina itens pelo nome do item (e nao pelo número da posição)
print('J)', preços)  #reparar que eliminou apenas o primeiro 88 (haviam 2)
if 400 in preços:        #removendo itens que talvez não existam na lista:
  preços.remove(400)
else:
  print('Não achei o número 400 para removê-lo')
preços.clear()       #esvazia toda a variável.
print('K)', preços)
preços = []          #esvazia toda a variável (como o .clear())
print('L)', preços)   
del preços    #del nao é um método, e sim um comando (métodos recebem parenteses no final)
```

```
#COPIAR LISTAS:
#CUIDADO há diferença entre copiar uma lista e criar uma ligação entre listas:
a = [2, 3, 4, 7]
b = a              #aqui foi criada uma LIGAÇÃO entre as listas A e B, que será válida em todos os pontos do programa daqui em diante
b[2] = 8
print(f'Lista A: {a}')    #não foi feita uma cópia, e sim uma ligação!
print(f'Lista B: {b}')

print('\n*****\n')

#COPIANDO UMA LISTA:
c = [1, 2, 3, 4]
d = c[:]   #aqui foi criada uma CÓPIA: todos os valores em C serão inseridos em D
d[3] = 14
print(f'Lista C: {c}')    #não foi feita uma cópia, e sim uma ligação!
print(f'Lista D: {d}')
```

Desenvolva um programa em que o usuário possa digitar cinco valores numéricos e cadastre-os em uma lista, JÁ NA POSIÇÃO CORRETA (sem usar o sort() no final). No final, mostre a lista ordenada na tela:

https://www.youtube.com/watch?v=QDpwjBYRcVE&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=11

```
lista3 = list()
maior = menor = 0
for c in range (0, 5):
  num = int(input('Digite um valor: '))
  if c == 0: 
    lista3.append(num)      #o primeriro num será colocado diretamente na lista
    print(f'Item {num} adicionado na lista com sucesso.')
  elif num > lista3[-1]:    #se o num for maior que o ultimo item da lista, ele será colocado no final da lista
    lista3.append(num)      #como as linhas 6 e 8 são iguais, c comando poderia ser simplificado (incluindo a linha 7 na linha 5): if c == 0 or num < lista3[0]
    print(f'Item {num} adicionado no final da lista com sucesso.')
  else:
    pos = 0
    while pos < len(lista3):   #o while terá tantos laços quanto o tamanho da lista
      if num <= lista3[pos]:   #em cada laço, vai verificar se o num é maior ou igual ao item da lista que estiver na posição do pos
        lista3.insert(pos, num)
        print(f'Item {num} adicionado na lista com sucesso, na posição {pos}.') 
        break
      pos += 1
print(f'Os valores digitados, em ordem, foram: {lista3}')
```

# EXERCÍCIO 083 - Validando expressões matemáticas:

Desenvolva um programa em que o usuário digite uma expressão qualquer que use parênteses. Seu aplicativo deverá analisar se a expressão está com parenteses abertos e fechados na ordem correta.

https://www.youtube.com/watch?v=dvhP41Z7TLk&list=PLHz_AreHm4dksnH2jVTIVNviIMBVYyFnH&index=14



```
expressão = str(input('Digite uma expressão matemática: '))     #é uma variável simples
pilha = []                   #lista em que vamos separar os parenteses abrindo "("
for símbolo in expressão:    #o for vai "rodar" para cada caractere da variável
  if símbolo == '(':
    pilha.append('(')
  elif símbolo == ')':
    if len(pilha) > 0:       #caso haja um parenteses abrindo na lista
      pilha.pop()            #apaga o último item da lista, que será um "("
    else:
      pilha.append(')')
      break
if len(pilha) == 0:
  print('Sua expressão está válida.')
else:
  print('Sua expressão está errada.')
```

# EXERCÍCIO 088 - Palpites para a mega sena:

Desenvolva um programa que ajude um jogador da Mega Sena a criar palpites. O programa vai perguntar quantos jogos serão gerados e vai sortear 6 números entre 1 e 60 para cada jogo, cadastrando tudo em uma lista composta.

https://www.youtube.com/watch?v=Hd7Ycaj61xE

```
from random import randint
palpites = list()
jogo = list()
qtde = num = numjogo = 0
print('=-' * 25)
print('{:^50}'.format('GERADOR DE PALPITES PARA MEGA SENA:'))
print('=-' * 25)
qtde = int(input('Quandos palpites serão gerados?: '))
for c in range(0, qtde):
  contador = 0    #conta para que um jogo tenha 6 números
  while True:
    num = randint(1, 60)
    if num not in jogo:
      jogo.append(num)
      contador += 1
    if contador == 6:
      break
  jogo.sort()
  print(f'O palpite para o {numjogo + 1}º jogo é: {jogo}')
  palpites.append(jogo[:])
  numjogo += 1
  jogo.clear()

print(f'\n Registro de palpites: {palpites}')
```

Desenvolva um programa que leia o nome e duas notas de vários alunos e guarde tudo em uma lista composta. No final, mostre um boletim contendo a média de cada um e permita que o usuário possa mostrar as notas de cada aluno individualmente:

https://www.youtube.com/watch?v=7xrCJnniqMw

```
#1) guardando os dados em uma lista composta:
classe = []
aluno = []
notas = []
while True:
  aluno.append(str(input('Nome: ')).strip().upper())
  notas.append(int(input('Nota 1: '))) 
  notas.append(int(input('Nota 2: ')))
  aluno.append(notas[:])
  classe.append(aluno[:])
  notas.clear()
  aluno.clear()
  média = 0
  continuar = ' '
  continuar = str(input('Deseja continuar? [S/N]: ')).strip().upper()[0]
  if continuar == 'N':
    break
print(classe)

#2)Formatando e apresentando tabela com as médias dos alunos:
print('=-'*15)
print('{:^30}'.format('Nº   NOME          MÉDIA'))
print('=-'*15)
for pos, item in enumerate(classe):
  print(f'{pos:^7}', end = ' ')
  print(f'{classe[pos][0]:<14}', end = ' ')
  média = (classe[pos][1][0] + classe[pos][1][1]) / 2
  print(f'{média}')
print('=-'*15)

#3)Mostrando as notas dos alunos separadamente:
while True:
  resposta = ' '
  resposta = int(input('Deseja ver as notas de qual aluno? [999 para parar] '))
  if resposta == 999:
    break
  print(classe[resposta])
print('>>Finalizando..<<\nVolte sempre.lk')
```
