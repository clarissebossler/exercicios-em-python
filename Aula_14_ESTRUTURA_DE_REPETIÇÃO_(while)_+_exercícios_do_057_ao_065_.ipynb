#ESTRUTURA DE REPETIÇÃO (while)

*   Não é necessário determinar o número de repetições (range), como o FOR.
*   O que deve ser determinado então é uma condição para encerrar a repetição.
*   Se chama "Estrutura de repetição com teste lógico"
*   O while é mais versátil que o for.
*   É melhor utilizar o for quando o limite é conhecido.
*   Vantagem do while sobre o for: não está preso a um padrão (ver figura 2)

#  EXERCÍCIO 058 - Jogo da adivinhação v.2.0. (guessing game):
Melhore o jogo desenvolvido na instrução 028 em que "o computador pensa" em um número de 0 a 10, e o usuário tenta adivinhar. Agora o jogador tenta adivinhar até acertar, e o programa informa quantas tentativas foram necessárias:
 
Improve game developed at instruction 028, when "computer thinks" of a 0 - 10 number and the user tries to guess. But now, user tries until get right, and program inform him how many gesses he nedded:
 
https://www.youtube.com/watch?v=-QkOIHJ1Chw&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=27

```
tentativas = 1
jogador = int(input('Vou jogar um dado D6... Qual número você acha que vai sair? '))
from random import randint
dado = randint(1, 6)
while jogador != dado:
  jogador = int(input('Não foi esse! Tente mais uma vez: '))
  tentativas += 1
print('Você acertou! Meu dado saiu o número {}. Você precisou de {} palpite(s) para adivinhar.'. format(dado, tentativas), end = ' ')
if tentativas == 1:
  print('Você é bom!')
```

```
#Outra solução, com true/ false:
tentativas = 0            #Aqui ficou 0 (e nao 1 como no primeiro exemplo, porque a variável jogador já está dentro do while (parece melhor))
acertou = False
print('Vou jogar um dado D10... Tente adivinhar o número que saiu! ')
from random import randint
dado = randint(1, 10)
while not acertou:
    jogador = int(input('Qual é o seu palpite? '))
    tentativas += 1
    if jogador == dado:
      acertou = True
    else:
      if dado > jogador:
        print('Foi mais do que isso...')
      elif dado < jogador:
        print('Foi menos do que isso...')
print('Sim, saiu {}. Você acertou com {} tentativa(s).'.format(dado, tentativas), end = ' ')
if tentativas <= 3:
  print('Você é bom!')
```

#  EXERCÍCIO 059 - Menu de oções (options menu):
Desenvolva um programa que leia dois valores e mostre um menu com as opções: SOMAR, MULTIPLICAR, MAIOR, NOVOS NÚMEROS, SAIR DO PROGRAMA. O programa deverá realizar a operação solicitada em cada caso:
 
Develop a program that read two numbers and shows the following options: ADD, MULTIPLY, NEW NUMBERS, EXIT. Program must perform the requested operation:
 
https://www.youtube.com/watch?v=OBJL5vPj4-E&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=28

```
opção = 0
n1 = int(input('Primeiro número: '))
n2 = int(input('Segundo número: '))
while opção is not 5:
  print('.=.' * 12)
  print('Qual operação você deseja realizar?')
  opção = int(input('''  [1] SOMAR
  [2] MULTIPLICAR
  [3] MAIOR
  [4] NOVOS NÚMEROS
  [5] SAIR DO PROGRAMA '''))
  if opção == 1:
    soma = n1 + n2
    print('A soma {} + {} resulta em {}'.format(n1, n2, soma))
  elif opção == 2:
    produto = n1 * n2
    print('O resultado de {} x {} é {}'.format(n1, n2, produto))
  elif opção == 3:
    if n1 > n2:
      print('O maior número é o {}'.format(n1))
    elif n2 > n1:
      print('O maior numero é o {}'.format(n2))
    else:
      print('Os dois números são iguais.')
  elif opção == 4:
    n1 = int(input('Novo primeiro número: '))
    n2 = int(input('Novo segundo número: '))
  elif opção == 5:
    print('Finalizando', end = '')
    from time import sleep
    for c in range (0, 3):
      sleep(1)
      print('.', end = '')
  else:
    print('Opção inválida.')
print('\n')
print('.=.' * 12)
```

https://www.youtube.com/watch?v=w7yn1_Mfu0E&list=PLHz_AreHm4dk_nZHmxxf_J0WRAqy5Czye&index=32
